```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.formula.api import glm
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
df = pd.read_csv('/content/EC5003 PC Wk 4 Data.csv')
df.head()
df.describe()
```

```python
np.log(df['Q'])
```

```python
df['LnPB']=np.log(df['PB'])
```

```python
df.head()
df.describe()
```

```python
df['LnPL']=np.log(df['PL'])
df['LnPR']=np.log(df['PR'])
df['LnI']=np.log(df['I'])
df.head()
df.describe()
```

```python
mod=ols('LnQ~LnPB+LnPL+LnPR+LnI',data=df).fit()
print(mod.summary())
```

```python
hypotheses = '(LnPB+LnPL+LnPR+LnI=0)'
f_test = mod.f_test(hypotheses)
print(f_test)
```

The test statistic is the same as with Gretl as if the associated p value - unable to reject the null hypothesis.

```python
mod=ols('LnQ~LnPB+LnPL+LnPR+LnI',data=df).fit()
hypotheses = '(LnPB+LnPL+LnPR+LnI=0)'
f_test = mod.f_test(hypotheses)

#In order to impose non-exclusion linear restrictions we need to use glm
#Swapping ols with glm we can get the same results as ols
mod = glm('LnQ~LnPB+LnPL+LnPR+LnI',data=df)

#But now we can impose restrictions
mod_rls = mod.fit_constrained(hypotheses)
print(mod_rls.summary().tables[1])
print('aic and bic',mod_rls.aic,mod_rls.bic_llf)
```

This table of results show the model with the restriction imposed which is the same result as given by Gretl
