Need to call up the required autocorrelation test packages from statsmodels

```python
from statsmodels.stats.diagnostic import acorr_breusch_godfrey as bgtest
from statsmodels.stats.stattools import durbin_watson
```

```python
df = pd.read_csv('/content/Price Wk6.csv')
df.head()
df.describe()
```

```python
mod=ols('Q~P',data=df).fit()
print(mod.summary())
```

```python
[durbin_watson(mod.resid),.5*(2-durbin_watson(mod.resid))]
```

You can see the DW test stat is the same as produced in the table of results

Next, we show the BG test

```python
bg1=bgtest(mod,nlags=1)
print('\nbreusch godfrey:','LM',bg1[0],'LM-pval',bg1[1],': F',bg1[2],'F-pval',bg1[3])
```
