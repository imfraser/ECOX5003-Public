```python
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
import matplotlib as plt
```

```python
df = pd.read_csv('/content/EC5003 PC Wk 1 US Wage Data April 2025.csv')
df.head()
df.describe()
```

```python
mod=ols('Lnwage~educ+exper+tenure',data=df).fit()
print(mod.summary())
```

```python
mod2=ols('Lnwage~educ+exper+expten',data=df).fit()
print(mod2.summary())
```

```python
mod3=ols('Lnwage~educ+exper+exper2+female',data=df).fit()
print(mod3.summary())
```
